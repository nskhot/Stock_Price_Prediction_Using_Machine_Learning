# Replacing Null Values

```python
data['Open'].fillna(data['Open'].mean(), inplace= True)
data['High'].fillna(data['High'].mean(), inplace= True)
data['Low'].fillna(data['Low'].mode() [0], inplace= True)
data['Close'].fillna(data['Close'].mode() [0], inplace= True)
data['Volume'].fillna(data['Volume'].mean(), inplace= True)
data['Adj Close'].fillna(data['Adj Close'].mean(), inplace= True)
```

```python
data.isnull().sum()
```

```python
#data.index= data["Date"]
#data
```

```python
data.rename(columns = {"Adj Close" : "Adj_Close"},inplace = True)
data
```

```python
data =data[["Adj_Close"]]
data.head()
```

```python
data["Prediction"]=data["Adj_Close"].shift(-1)
data.head()
```

```python
data.tail()
```

```python
x= np.array(data["Adj_Close"])
x=x[:-1]
x=pd.DataFrame(x)
x
```

```python
data["Adj_Close"].plot()
```

```python
y=np.array(data["Prediction"])
y=pd.DataFrame(y)
y
```

```python
y=y[:-1]
y
```

# Perform Train_Test Splittiing

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2)
```

```python
xtrain.shape,xtest.shape
```

```python
ytrain.shape,ytest.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
LR.fit(xtrain,ytrain)
```

```python
pred= LR.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
ac= LR.score(pred,ytest)
print("Accuracy Score is :",ac *100)

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error 

mae = mean_absolute_error(ytest,pred)
print("Error: ",mae)

mse = mean_squared_error(ytest, pred)
print("Error:", mse)


```
